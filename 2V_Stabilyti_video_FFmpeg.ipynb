### Этот код должен стабилизировть все видео в папке

```python
import os
import subprocess
import time
import re
from tqdm.notebook import tqdm
import psutil
import GPUtil
from datetime import timedelta
import torch

def get_gpu_usage():
    gpus = GPUtil.getGPUs()
    if gpus:
        gpu = gpus[0]
        return f"GPU использование: {gpu.load*100:.1f}%, Память: {gpu.memoryUsed}/{gpu.memoryTotal} MB"
    return "GPU не доступен"

def get_video_duration(input_path):
    cmd = ['ffprobe', '-v', 'error', '-show_entries', 'format=duration', '-of', 'default=noprint_wrappers=1:nokey=1', input_path]
    result = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, universal_newlines=True)
    return float(result.stdout)

def get_unique_filename(base_path):
    directory, filename = os.path.split(base_path)
    name, ext = os.path.splitext(filename)
    counter = 1
    while os.path.exists(base_path):
        base_path = os.path.join(directory, f"{name}_{counter}{ext}")
        counter += 1
    return base_path

def check_cuda_availability():
    return torch.cuda.is_available()

def stabilize_video(input_path, output_path):
    try:
        print(f"\nОбработка видео: {os.path.basename(input_path)}")
        print(f"Текущее {get_gpu_usage()}")

        video_duration = get_video_duration(input_path)
        print(f"Длительность видео: {video_duration:.2f} секунд")

        cuda_available = check_cuda_availability()
        print("CUDA доступна. Используем GPU для ускорения." if cuda_available else "CUDA недоступна. Используем CPU.")

        # Анализ движения
        analyze_cmd = [
            'ffmpeg',
            '-hwaccel', 'cuda' if cuda_available else 'auto',
            '-i', input_path,
            '-vf', 'vidstabdetect=stepsize=6:shakiness=8:accuracy=15:result=transforms.trf',
            '-f', 'null', '-'
        ]

        print("Начало анализа движения...")
        start_time = time.time()
        process = subprocess.Popen(analyze_cmd, stderr=subprocess.PIPE, universal_newlines=True)

        pbar_analyze = tqdm(total=100, desc="Анализ", unit="%")
        for line in process.stderr:
            if 'time=' in line:
                time_match = re.search(r'time=(\d{2}):(\d{2}):(\d{2}\.\d{2})', line)
                if time_match:
                    current_time = (int(time_match.group(1)) * 3600 +
                                    int(time_match.group(2)) * 60 +
                                    float(time_match.group(3)))
                    progress = min(int((current_time / video_duration) * 100), 100)
                    pbar_analyze.update(progress - pbar_analyze.n)

        process.wait()
        pbar_analyze.close()
        analyze_time = time.time() - start_time
        print(f"\nАнализ движения завершен за {analyze_time:.2f} секунд")
        print(f"Текущее {get_gpu_usage()}")

        # Стабилизация видео
        stabilize_cmd = [
            'ffmpeg',
            '-hwaccel', 'cuda' if cuda_available else 'auto',
            '-i', input_path,
            '-vf', 'vidstabtransform=input=transforms.trf:zoom=1:smoothing=30,unsharp=5:5:0.8:3:3:0.4',
            '-c:v', 'h264_nvenc' if cuda_available else 'libx264',
            '-preset', 'p4' if cuda_available else 'medium',
            '-cq', '23',
            '-c:a', 'copy',
            output_path
        ]

        print("Начало процесса стабилизации...")
        start_time = time.time()
        process = subprocess.Popen(stabilize_cmd, stderr=subprocess.PIPE, universal_newlines=True)

        pbar_stabilize = tqdm(total=100, desc="Стабилизация", unit="%")
        for line in process.stderr:
            if 'time=' in line:
                time_match = re.search(r'time=(\d{2}):(\d{2}):(\d{2}\.\d{2})', line)
                if time_match:
                    current_time = (int(time_match.group(1)) * 3600 +
                                    int(time_match.group(2)) * 60 +
                                    float(time_match.group(3)))
                    progress = min(int((current_time / video_duration) * 100), 100)
                    pbar_stabilize.update(progress - pbar_stabilize.n)

        process.wait()
        pbar_stabilize.close()
        stabilize_time = time.time() - start_time
        print(f"\nСтабилизация завершена за {stabilize_time:.2f} секунд")
        print(f"Текущее {get_gpu_usage()}")

        total_time = analyze_time + stabilize_time
        print(f"Общее время обработки: {total_time:.2f} секунд")
        return True, total_time
    except subprocess.CalledProcessError as e:
        print(f"Ошибка при стабилизации видео: {e}")
        return False, 0
    finally:
        if os.path.exists('transforms.trf'):
            os.remove('transforms.trf')

def get_stabilized_videos(output_folder):
    stabilized_videos = {}
    for f in os.listdir(output_folder):
        if f.startswith('stabilized_'):
            original_name = re.sub(r'stabilized_(.+?)(_\d+)?\.[^.]+$', r'\1', f)
            extension = os.path.splitext(f)[1]
            full_original_name = f"{original_name}{extension}"

            if full_original_name not in stabilized_videos or os.path.getsize(os.path.join(output_folder, f)) > stabilized_videos[full_original_name]:
                stabilized_videos[full_original_name] = os.path.getsize(os.path.join(output_folder, f))
    return stabilized_videos

def process_videos_in_folder(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    video_extensions = ('.mp4', '.avi', '.mov', '.mkv')
    all_videos = [f for f in os.listdir(input_folder) if f.lower().endswith(video_extensions)]
    stabilized_videos = get_stabilized_videos(output_folder)

    videos_to_process = []
    for video in all_videos:
        input_path = os.path.join(input_folder, video)
        input_size = os.path.getsize(input_path)

        if video in stabilized_videos:
            stabilized_size = stabilized_videos[video]
            if stabilized_size >= input_size * 0.3:  # Проверка на 30% от исходного размера
                print(f"Видео {video} уже стабилизировано (размер: {stabilized_size / (1024*1024):.2f} МБ). Пропускаем.")
                continue
            else:
                print(f"Размер стабилизированного видео {video} меньше 30% от исходного. Перестабилизируем.")
        else:
            print(f"Видео {video} не найдено среди стабилизированных. Будет обработано.")

        videos_to_process.append(video)

    print(f"Найдено {len(all_videos)} видеофайлов в папке {input_folder}")
    print(f"Из них {len(videos_to_process)} требуют обработки")

    for video_file in videos_to_process:
        input_path = os.path.join(input_folder, video_file)
        output_path = get_unique_filename(os.path.join(output_folder, f"stabilized_{video_file}"))

        print(f"\nОбработка файла: {video_file}")
        success, processing_time = stabilize_video(input_path, output_path)

        if success:
            print(f"Стабилизация успешно завершена.")
            print(f"Время обработки: {processing_time:.2f} секунд")
            if os.path.exists(output_path):
                print(f"Стабилизированное видео создано. Размер: {os.path.getsize(output_path) / (1024*1024):.2f} МБ")
            else:
                print("Стабилизированное видео не было создано.")
        else:
            print("Произошла ошибка при стабилизации видео.")

    print("\nОбработка всех видео завершена.")

# Основной блок выполнения
if __name__ == "__main__":
    input_folder = input("Введите путь к папке с исходными видео: ")
    output_folder = input("Введите путь к папке для сохранения стабилизированных видео (или нажмите Enter для использования той же папки): ")

    if not output_folder:
        output_folder = os.path.join(input_folder, "stabilized")

    if not os.path.exists(input_folder):
        print("Указанная входная папка не существует.")
    elif not os.access(input_folder, os.R_OK):
        print("Нет прав на чтение из входной папки.")
    elif not os.access(os.path.dirname(output_folder), os.W_OK):
        print("Нет прав на запись в выходную папку.")
    else:
        process_videos_in_folder(input_folder, output_folder)
```
